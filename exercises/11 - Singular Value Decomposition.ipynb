Let us consider $N$ data points $\{x_k\}, k=1, ..., N$ in $\mathbb{R}^d$. We will denote by $X$ the $N\times d$ matrix s.t. :

$
X = \begin{pmatrix}
x_1^T\\
x_2^T\\ \vdots \\ x_N^T \end{pmatrix}$

The sample covariance matrix is defined by $C = \frac{1}{N}X^TX$.
